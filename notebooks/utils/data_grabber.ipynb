```python
def get_file(url, out):
    response = requests.get(url)
    totalbits = 0
    if response.status_code == 200:
        with open(out, 'wb') as f:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    totalbits += 1024
                    f.write(chunk)
    print(f"Downloaded from {url} file = ",totalbits*1025,"KB...")
```

### Process EIA spreadsheets and convert to .csv for reading by other processes

```python
EIA_DL_URL = 'http://ir.eia.gov/wpsr/psw09.xls'
EIA_OUTFILE= '../data/psw09.xls'

# Fetch the xls file
get_file(EIA_DL_URL, EIA_OUTFILE)

# Pipeline data conversion to .csv
df = pd.read_excel(open(EIA_OUTFILE, 'rb'),
                   skiprows=1,
                   sheet_name='Data 6'
                   #converters= {'Date': pd.to_datetime},
                   #index_col='Date'
                  )
df = df.drop([0]) #remove desc
df.rename(columns={'Sourcekey':'Date'},inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

file_name = "../data/all_eia_stock_sheet_latest.csv"
df.to_csv(file_name, sep=',', encoding='utf-8')
df.tail(5)
```

```python
EIA_OIL_PRICE_URL = 'http://www.eia.gov/dnav/pet/hist_xls/RCLC1w.xls'
EIA_OIL_PRICE_OUTFILE = '../data/RCLC1w.xls'

get_file(EIA_OIL_PRICE_URL, EIA_OIL_PRICE_OUTFILE )

df = pd.read_excel(open(EIA_OIL_PRICE_OUTFILE, 'rb'),
                   skiprows=1,
                   sheet_name='Data 1'
                   #converters= {'Date': pd.to_datetime},
                   #index_col='Date'
                  )
df = df.drop([0]) #remove desc
df.rename(columns={'Sourcekey':'Date'},inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

file_name = "../data/eia_pricing_latest.csv"
df.to_csv(file_name, sep=',', encoding='utf-8')
df.tail(5)
```

```python
EIA_OIL_FUTURES = 'https://www.eia.gov/dnav/pet/xls/PET_PRI_FUT_S1_W.xls'
EIA_OIL_FUTURES_OUTFILE = '../data/futures.xls'

get_file(EIA_OIL_FUTURES, EIA_OIL_FUTURES_OUTFILE)

df = pd.read_excel(open(EIA_OIL_FUTURES_OUTFILE, 'rb'),
                   skiprows=1,
                   sheet_name='Data 1'
                   #converters= {'Date': pd.to_datetime},
                   #index_col='Date'
                  )
df = df.drop([0]) #remove desc
df.rename(columns={'Sourcekey':'Date'},inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

file_name = "../data/eia_futures_latest.csv"
df.to_csv(file_name, sep=',', encoding='utf-8')
df.tail(5)
```
